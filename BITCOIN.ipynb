# DATA ANALYSIS ON BITCOIN HISTORICAL DATA
## Importing the necessary modules

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Reading the data file containing data on the bitcoin historical data.

bitcoin_data = pd.read_csv("bitcoin.csv")
bitcoin_data
```

```python
#Checking the number of rows and columns in the dataset.

bitcoin_data.shape
```

## NULL VALUES

### Checking for null values
The sections that return True indicate that there is a null value

```python
bitcoin_data.isnull()
```

```python
#Establishing the number of rows that contain null values.

bitcoin_data.isnull().sum()
```

```python
#Removing null values

bitcoin_data.dropna(inplace = True)
bitcoin_data.isnull().sum()
```

```python
bitcoin_data
```

```python
#using the info to get information of the data.

bitcoin_data.info()
```

### DATA VISUALIZATION

```python
bitcoin_data.head()
```

```python
x = bitcoin_data["Timestamp"]
y = bitcoin_data["Volume_(Currency)"]

plt.plot(x, y)
plt.title("Visualizing")
plt.xlabel("Timestamp")
plt.ylabel("Volume_Currency")

plt.show()
```

```python
x = bitcoin_data["Timestamp"]
y = bitcoin_data["Open"]

plt.plot(x, y)

plt.xlabel("Timestamp")
plt.ylabel("Open price")

plt.show()
```

```python
x = bitcoin_data["Volume_(BTC)"]
y = bitcoin_data["High"]

plt.plot(x, y)

plt.xlabel("Volume")
plt.ylabel("High")

plt.show()
```
